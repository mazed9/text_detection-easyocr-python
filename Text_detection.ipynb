###Read image.

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/Text detection/data/test_image.jpg")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

This is the image we're gonna detect text on.

###Instance text detector

```python
reader=easyocr.Reader(['en'],gpu=False)
```

###Detect text

```python
text_=reader.readtext(img)
```

Let's see what we've got in "text_".

```python
for t in text_:
  print(t)
```

For each element(text) in text_ we get a tuple. The tuple has a list containig the coordinates of the corners of the bounding box (top-left, bottom-left, bottom-right, top-right), the text and the confidence score.

###Draw bounding box and text.

```python
for t in text_:
  bbox, text, score =t
  cv2.rectangle(img, bbox[0], bbox[2], (0,255,0), 2)
  cv2.putText(img, text, bbox[0], cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)
```
